```python
import codecs,nltk,string
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.__version__
```

```python
%matplotlib inline
```

```python
from nltk.corpus import PlaintextCorpusReader
corpusFiles = PlaintextCorpusReader("/Users/gilles/Github/Zipf", '.*\.txt')
corpusFiles.fileids()
```

```python
mots=[w for w in corpusFiles.words('Swan.txt') if not w in string.punctuation]
```

```python
words = pd.DataFrame(mots, columns=['word'])
word_counts = words.word.value_counts().reset_index()
word_counts.columns = ['word', 'n']
word_counts['word_rank'] = word_counts.n.rank(ascending=False)
```

```python
import seaborn as sns
sns.set_context("paper")
f, ax = plt.subplots(figsize=(50, 5))
ax.set(xscale="linear", yscale="linear")
sns.distplot(word_counts.set_index("n")["word_rank"])
```

```python
sns.set_context("poster")
f, ax = plt.subplots(figsize=(7, 7))
ax.set(xscale="log", yscale="log")
sns.regplot("n","word_rank", word_counts, ax=ax, scatter_kws={"s": 50})
```

```python
wordFreqs={}
for word in corpusFiles.words('Swan.txt'):
    if not word in string.punctuation:
        wordLower=word.lower()
        if not wordLower in wordFreqs:
            wordFreqs[wordLower]=0
        wordFreqs[wordLower]+=1
```

```python
dfFreqs=pd.DataFrame.from_dict(wordFreqs,orient="index")#.sort_values("0")
dfFreqs.columns=["freq"]
dfFreqs["rank"]=dfFreqs.freq.rank(ascending=False)
```

```python
dfFreqs.sort_values("rank",ascending=True,inplace=True)
```
